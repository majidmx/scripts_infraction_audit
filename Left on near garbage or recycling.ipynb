```python

import pandas as pd
import numpy as np

df=pd.read_csv(r'C:\Users\majidmm\Documents\MTI\verified_allcol.csv')
df=df[df['root_cause'].str.contains('left on/near garbage or recycling')==True]

df['Data']=df['Data'].str.lower() 
df['Infraction_Data']=df['Infraction_Data'].str.lower() 
df['Additional_Details_Data']=df['Additional_Details_Data'].str.lower()

word_list_1=('left','dump','put','deliver')

word_list_2=('dumpster','garbage','trash','recycling bin')


df['attribute_1'] = df.Additional_Details_Data.str.extract('({})'.format('|'.join(word_list_1)),expand=False).str.lower()
df['attribute_2'] = df.Additional_Details_Data.str.extract('({})'.format('|'.join(word_list_2)),expand=False).str.lower()

df['Second_Status']=(df['Additional_Details_Data'].str.contains('left|dump|put|deliver'))&(df['Additional_Details_Data'].str.contains('trash|dumpster|garbage|recycling bin'))

df.to_csv(r'C:\Users\majidmm\Documents\MTI\verified_near garbage or recycling.csv',columns=('ShortId','tid','root_cause','Infraction','Additional Details','First_Status','Second_Status'),index = False)


df


```
