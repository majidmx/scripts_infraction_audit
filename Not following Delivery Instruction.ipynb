```python
import pandas as pd
import numpy as np

df=pd.read_csv(r'C:\Users\majidmm\Documents\MTI and RC\SAMPLE AUDIT\OUTPUT_FILES\verified_infractions_first_check.csv')
df=df[df['root_cause'].str.contains('not following delivery instruction')==True]

df['Data']=df['Data'].str.lower() 
df['Infraction']=df['Infraction'].str.lower() 
df['Additional_Details']=df['Additional_Details'].str.lower()

df1=pd.read_csv(r'C:\Users\majidmm\Documents\MTI and RC\SAMPLE AUDIT\INPUT_FILES\comp_data.csv',encoding='ISO-8859-1')
df1.dropna(axis = 0, how = 'all', inplace = True) 
df2=df.merge(df1,left_on='tid',right_on='Tracking Id',how='left')

df2['Customer Notes']=df2['Customer Notes'].str.lower()
df2['Reason']=df2['Reason'].str.lower()


word_list=['front door only','leave safe in mailroom','leave safe in customer','leave safe in locker','mail room','leave safe in front door','leave safe in rear door','leave safe in receptionist']
word_list_1=['front door only','leave safe in mailroom','leave safe in customer','locker','mail room','front door','rear door','receptionist']

df2['Attribute'] = df2['Customer Notes'].str.extract('({})'.format('|'.join(word_list)), 
                        expand=False).str.lower()


df2['Attribute_1'] = df2['Attribute'].str.extract('({})'.format('|'.join(word_list_1)), 
                        expand=False).str.lower()

df2['Attribute_1'] = df2['Attribute_1'].fillna('No data available')
df2['Attribute_1'] = df2['Attribute_1'].astype(str)

df2['Attribute_1'] = df2['Attribute_1'].replace('leave safe in customer', "receiver") 
df2['Attribute_1'] = df2['Attribute_1'].replace('leave safe in mailroom', "mail room") 
df2['Attribute_1'] = df2['Attribute_1'].replace('front door only', "front door") 


df2['Second_Status']=np.where(df2['Reason']==df2['Attribute_1'],'True','False')


df2.to_csv(r'C:\Users\majidmm\Documents\MTI and RC\SAMPLE AUDIT\OUTPUT_FILES\verified_NfDI_1stcheck.csv',index = False)


df2



```
