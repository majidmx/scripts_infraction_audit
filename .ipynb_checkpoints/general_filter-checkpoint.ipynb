```python
df2=pd.read_csv(r'C:\Users\majidmm\Documents\MTI\Combined.csv',usecols=('ShortId','Data'))
df2['ShortId']=df2['ShortId'].astype(str)
df2_1=pd.read_csv(r'C:\Users\majidmm\Documents\MTI\query.csv',usecols=('case_id','details','root_cause','tid'),encoding='ISO-8859-1')

df2_2 = df2_1.rename(columns = {"case_id": "ShortId"}) 

df2_2['ShortId'].dropna(inplace=True)


x=df2_2.merge(df2,on='ShortId',how='left')
x['Data'] = x.Data.fillna('No data available')

# x = x[(x['root_cause'] != 'Positive Feedback')&(x['root_cause'] != 'Offboarding - Combined Threshold')]
x['Infraction_Data']=x['Data'].apply(lambda st: st[st.find("Infraction:")+11:st.find("Additional")])
x['Additional_Details_Data']=x['Data'].apply(lambda st: st[st.find("Additional Details"):st.find("Delivery Tip")])


x=x[x['Data'].str.contains('Appeal Successful|Appeal Rejected|Overturned during MTI process|DSP DA station tickets are now processed through RIVER')==False]

x.to_csv(r'C:\Users\majidmm\Documents\MTI\all_infractions_data.csv',index=False)
x
```

```python
import numpy as np


df3=pd.read_csv(r'C:\Users\majidmm\Documents\MTI\all_infractions_data.csv')
# df3 = df3[(df3['root_cause'] != 'Duplicate')& (df3['root_cause'] != 'Not a Tier Infraction')&(df3['root_cause'] != 'Positive Feedback')&(df3['root_cause'] != 'Offboarding - Combined Threshold')]

df3['Infraction']=df3['Data'].apply(lambda st: st[st.find("Infraction:")+11:st.find("Additional")])
df3['Additional Details']=df3['Data'].apply(lambda st: st[st.find("Additional Details"):st.find("Delivery Tip")])

df3['root_cause']=df3['root_cause'].str.lower() 
df3['Infraction']=df3['Infraction'].str.lower()



word_list=['badge','bags left','behavior towards a customer','business closed','dangerous','delivered to receiver','not following','did not follow sop','headphones','incorrect scan',
    'inside lobby','lawn','leaving keys','left in driveway','left in road','left unattended','locker','mail box','mailroom/leasing','no safe location',
    'non-swa','on/near garbage','ovesize','safety vest','seat belt','station employee','station standard work compliance','tailgating','throwing','unable to access',
    'unauthorized entry','under the roof','unsafe driving at station','property damage','unsecure packages']

df3['Infraction1'] = df3.root_cause.str.extract('({})'.format('|'.join(word_list)), 
                        expand=False).str.lower()

df3['Infraction2'] = df3.Infraction.str.extract('({})'.format('|'.join(word_list)), 
                        expand=False).str.lower()

df3['First_Status']=np.where(df3['Infraction1']==df3['Infraction2'],'True','False')

df4_1=df3[df3['Data'].str.contains('Appeal Successful|Appeal Rejected|Overturned during MTI process|DSP DA station tickets are now processed through RIVER')==False]
df4_1['Infraction'] = df4_1['Infraction'].str.strip()
df4_1['Data']=df4_1['Data'].str.lower() 

df4_1.to_csv(r'C:\Users\majidmm\Documents\MTI\verified_allcol.csv',index = False)

df4_1
```
