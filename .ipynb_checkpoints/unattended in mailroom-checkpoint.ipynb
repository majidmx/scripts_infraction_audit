```python
import pandas as pd
import numpy as np

df=pd.read_csv(r'C:\Users\majidmm\Documents\MTI\verified_allcol.csv')
df=df[df['root_cause'].str.contains('behavioral - left packages unattended in mailroom/leasing')==True]

df['Data']=df['Data'].str.lower() 
df['Infraction_Data']=df['Infraction_Data'].str.lower() 
df['Additional_Details_Data']=df['Additional_Details_Data'].str.lower()

word_list_1=('abandon','dump','outside','unattended','left')

word_list_2=('leasing','lobby','mail box','mail room','mail-room','mailbox','mailroom','parcel room')

df['attribute_1'] = df.Additional_Details_Data.str.extract('({})'.format('|'.join(word_list_1)), 
                        expand=False).str.lower()

df['attribute_2'] = df.Additional_Details_Data.str.extract('({})'.format('|'.join(word_list_2)), 
                        expand=False).str.lower()

df['Second_Status']=(df['Additional_Details_Data'].str.contains('deliver|left|unattended|dump|outside|abandon|leave|top'))&(df['Additional_Details_Data'].str.contains('office|leasing|mailroom|mail room|mailbox|mail box|parcel room|mail-room|lobby'))

df.to_csv(r'C:\Users\majidmm\Documents\MTI\verified_unattended_in_mailroom.csv',columns=('ShortId','tid','root_cause','Infraction','Additional Details','First_Status','Second_Status'),index = False)


df


```
