```python
df2=pd.read_csv(r'C:\Users\majidmm\Documents\MTI and RC\SAMPLE AUDIT\OUTPUT_FILES\verified_duplicate_t1.csv')
df2 = df2.rename(columns = {"inside_caseid":"case_id"}) 

df3=pd.read_csv(r'C:\Users\majidmm\Documents\MTI and RC\SAMPLE AUDIT\INPUT_FILES\audit_query.csv',usecols=('case_id','details','root_cause','tid'),encoding='ISO-8859-1')
# df3 = df3[df3['root_cause'] == 'Duplicate']


df2["case_id"] = pd.to_numeric(df2["case_id"],errors='coerce')
df3["case_id"] = pd.to_numeric(df3["case_id"],errors='coerce')

df4=df2.merge(df3,on='case_id',how='left')
# df6=pd.concat([df2, df3],join="inner")
df4['details_x']=df4['details_x'].astype(str)

df4['details_y']=df4['details_y'].astype(str)

df4['DAID_1']=df4['details_x'].apply(lambda st: st[st.find("(Transporter ID):")+17:st.find("Zip code of occurrence")])
df4['DAID_2']=df4['details_y'].apply(lambda st: st[st.find("(Transporter ID):")+17:st.find("Zip code of occurrence")])



df4['tid_Status']=np.where(df4['tid_x']==df4['tid_y'],'True','False')
df4['daid_Status']=np.where(df4['DAID_1']==df4['DAID_2'],'True','False')


df4.drop_duplicates(subset ="ShortId",keep = False, inplace = True) 


df4.to_csv(r'C:\Users\majidmm\Documents\MTI and RC\SAMPLE AUDIT\OUTPUT_FILES\verified_duplicate_t2.csv', index = False)
df4
```
