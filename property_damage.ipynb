```python
import pandas as pd
import numpy as np

df=pd.read_csv(r'C:\Users\majidmm\Documents\MTI\verified_allcol.csv')
df=df[df['root_cause'].str.contains('property damage')==True]

df['Data']=df['Data'].str.lower() 
df['Infraction_Data']=df['Infraction_Data'].str.lower() 
df['Additional_Details_Data']=df['Additional_Details_Data'].str.lower()

word_list=('damage','destroy','rut','hit','tear','tire','pulled off','stain','tore','leak','knock','dent','damaging','ditch','tracks','ruin','mess','mud','break','broke','stuck','drove all over','driver drove','driven into','scratch','grass','hole','torn','ran over','drove over')


df['attribute'] = df.Additional_Details_Data.str.extract('({})'.format('|'.join(word_list)), 
                        expand=False).str.lower()

df['Second_Status']=df['attribute'].str.contains('damage|destroy|rut|hit|tear|tire|pulled off|stain|tore|leak|knock|dent|damaging|ditch|tracks|ruin|mess|mud|break|broke|stuck|drove all over|driver drove|driven into|scratch|grass|hole|torn|ran over|drove over')

df.to_csv(r'C:\Users\majidmm\Documents\MTI\verified_property_damage.csv',columns=('ShortId','tid','root_cause','Infraction','Additional Details','First_Status','Second_Status'),index = False)


df


```
